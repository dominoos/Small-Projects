```python
import random



def start(): #will only run once to get player's name, maybe give context etc
  name= str(input("  GENERAL LIABILITY RELEASE FORM \n Sign the liability waiver: "))
  return name

def board(name, playerhealth, dealerhealth): #prints the board when stats change
  print("") # just for an extra line
  print("-----")
  print(name + ": " + str(playerhealth))
  print("Dealer:" + str(dealerhealth))
  print("-----\n")


def ammostats(live, blank): # just the read out at the start  hof the set.
  print(str(live) + " live rounds. " + str(blank) + " blanks.")


def bulletorder(live,blank): #this creates the order of the bullets
  print("Dealer inserts the shells in an unknown order")
  mag= []
  for i in range(live):
    mag.append(1)
  for i in range(blank):
    mag.append(0)
  random.shuffle(mag) #cool command which shuffles a list
  return mag

def playermove():
  print("Your move...")
  move= input("shoot yourself or shoot the Dealer? me, d? ").lower()
  while move != "me" and move != "d":
    move= input("CHOOOSEE.... ME OR D ")

  return 0 if move=="me" else 1 #0= player, 1= dealer


def dealermove():#idk code something cool here using percentages and stuff
  print("\nDEALERS MOVE...IT SHOOTS YOU")

  return(0) #for now always shoots player



def drama(who, mag, playerhealth, dealerhealth, sawedoff): #sawedoff should be 0 for false
  ignore=0 # REMOVE MAYBE
  print(who)
  h= [playerhealth, dealerhealth, ignore]
  h[who]-= (mag[0]+ sawedoff)

  print("blank." if mag[0]==0 else "LIVE!")
  if who==0 and mag[0]>0:
    sawedoff=0
    print("YOWWWWZAAA!!! YOU GOT SHOT!")
    board(name, h[0], h[1])
  elif who==1 and mag[0]>0:
    sawedoff=0
    print("DEALER GETS HIT!")
    board(name, h[0], h[1])
  return h[0], h[1]


def roundmatch(live, blank, ph, dh): # for round one
  print("\n\nNEW MATCH\n")
  board(name, ph,dh)

  ammostats(live,blank)
  mag= bulletorder(live,blank)
  turn=0
  while len(mag)>0:
    if turn%2==0:
      victim= playermove()
      if victim == 0:
        print("\nYou shoot yourself")
      else:
        print("\nYou shoot the Dealer")
    else:
      victim=dealermove()

    ph,dh= drama(victim, mag, ph, dh, 0)
    if victim==0 and mag[0]==0 and turn%2==0: #vic= player, blank bullet, players turn
      turn +=1
      print("Dealer's turn is skipped")
    mag.pop(0)


    turn+=1
    if ph<=0 or dh<=0:
      return ph,dh  #ends round early
  return ph, dh



def roundone(name):
  print("\nRound I ------------------------------------------")
  ph= 2 #player health
  dh=2 #dealer health
  ph,dh= roundmatch(1,2,ph,dh)
  ph,dh= roundmatch(3,2,ph,dh)
  if ph<=0:
    return 1
  elif dh<=0:
    return 2



```
