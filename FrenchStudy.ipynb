```python
#to update the sheets

def update_sheet(df):
  worksheet.update([df.columns.values.tolist()] + df.values.tolist())

update_sheet(df)
```

```python

def choosefrenchtestword(df):
  # Convert 'last seen' column to numeric, handling errors
  df["last seen"] = pd.to_numeric(df["last seen"], errors='coerce')
  sample= df.groupby("0French")["last seen"].min().sort_values()
  #display(sample)

  first_index= sample.index[0]
  testword= df.loc[df["0French"] == first_index, "0French"].iloc[0]
  #print(testword)

  return testword

#choosefrenchtestword(df)
```

```python

#Extract the first two columns and skip the first row
french_english_dict = dict(zip(df["0French"][0:], df["English"][0:]))

# Print the dictionary
print(french_english_dict)
```
